```python
from card import CardList, Card

cards_with_faces    = [card for card in data if "image_uris" not in card]
cards_without_faces = [card for card in data if "image_uris" in card]

models = CardList.model_validate(cards_without_faces)

for card in cards_with_faces:
    for face in card["card_faces"]:
        params = {
            "name": face["name"],
            "image_uris": face["image_uris"],
            "scryfall_uri": card["scryfall_uri"],
            "set_name": card["set_name"],
            "released_at": card["released_at"],
        }
        models.root.append(Card(**params,))
```
